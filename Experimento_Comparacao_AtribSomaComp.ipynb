## Experimentos
O experimento a seguir tem como o objetivo de analisar se há uma diferença no tempo de processamento quando se é feita a comparação entre dois valores, sejam eles variaveis. Assim, buscar uma resposta em caso da existencia de uma diferença, o quanto seria ela.

> Essa análise será realizada com a repetição de 1000 vezes o experimento, sendo cada trecho de código análisado (1 > 1, 1 < 1) repetidos 10^6 vezes para a obtenção de um valor analisável.

### Parâmetros do experimento

```python
r = 1000        # Número de repetições do experimento
# R = range(r)    # Vetor das repetições

n = 10 ** 6     # Quantidade do dado
# N = range(n)    # Vetor dos dados 'n'
```

### Código
> Experimento 1: A = 1

```python
# Define your experiment functions
def experiment_1():
    a = 1
        
# Create Experiment objects
experiment_1_obj = Experiment(repetitions=r, data_quantity=n)

# Run experiments
experiment_1_obj.run_experiment(experiment_1)
```

> Experimento 2: N == 1

```python
# Define your experiment functions
def experiment_2():
    n == 1

# Create Experiment objects
experiment_2_obj = Experiment(repetitions=r, data_quantity=n)

# Run experiments
experiment_2_obj.run_experiment(experiment_2)
```

> Experimento 3: A = N + 1

```python
# Define your experiment functions
def experiment_3():
    a = n + 1

# Create Experiment objects
experiment_3_obj = Experiment(repetitions=r, data_quantity=n)

# Run experiments
experiment_3_obj.run_experiment(experiment_3)
```

## Gráficos para demonstração resultados
### Gráfico em barras

```python
Experiment.show_bar_plot(experiment_1_obj, "Experiment 1")
Experiment.show_bar_plot(experiment_2_obj, "Experiment 2", "orange")
Experiment.show_bar_plot(experiment_3_obj, "Experiment 3", "red")
```

> Analisando os gráficos em barra nota-se que os tempos de ambos os experimentos se situam próximos a 0.045 e pela visualização é perceptível que há uma constancia nos valores obtidos, trazendo umta linha reta com poucos picos.

### Histograma

```python
# Show the histogram for Tc
Experiment.show_histogram(experiment_1_obj, "Experiment 1")

# Show the histogram for Tv
Experiment.show_histogram(experiment_2_obj, "Experiment 2", "orange")

# Show the histogram for Tv
Experiment.show_histogram(experiment_3_obj, "Experiment 3", "red")
```

> Com o histogram é possível de se observar que o experimento 2 possui seu valor de mais recorrencia a esquerda em relação ao valor de maior recorrencia do experimento 1, mostrando uma demora um pouco menor.
