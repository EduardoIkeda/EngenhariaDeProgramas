## Experimentos
O experimento a seguir tem como o objetivo de analisar se há uma diferença no tempo de processamento quando se é feita a comparação entre dois valores, sendo um deles costante e outro variável. Assim, buscar uma resposta em caso da existencia de uma diferença, o quanto seria ela. Para simplificar, está sendo feito o experimento apenas com a operação "Maior que".

> Essa análise será realizada com a repetição de 1000 vezes o experimento, sendo cada trecho de código análisado (1 > 1, 1 < 1) repetidos 10^6 vezes para a obtenção de um valor analisável.

### Parâmetros do experimento

```python
r = 1000        # Número de repetições do experimento
# R = range(r)    # Vetor das repetições

n = 10 ** 6     # Quantidade do dado
# N = range(n)    # Vetor dos dados 'n'
```

### Código
> Experimento 1: Constante > Constante

```python
# Define your experiment functions
def experiment_1():
    n > n
        
# Create Experiment objects
experiment_1_obj = Experiment(repetitions=r, data_quantity=n)

# Run experiments
experiment_1_obj.run_experiment(experiment_1)
```

> Experimento 2: Variável > Variável

```python
# Define your experiment functions
def experiment_2():
    1 > 1

# Create Experiment objects
experiment_2_obj = Experiment(repetitions=r, data_quantity=n)

# Run experiments
experiment_2_obj.run_experiment(experiment_2)
```

## Gráficos para demonstração resultados
### Gráfico em barras

```python
Experiment.show_bar_plot(experiment_1_obj, "Experiment 1")
Experiment.show_bar_plot(experiment_2_obj, "Experiment 2", "orange")
```

> Analisando os gráficos em barra nota-se que os tempos dos expeimentos são diferentes, o primeiro se situando próximo a 0.045 e o segundo próximo a 0.04. Ainda, pela visualização de ambos é perceptível que há uma constancia nos valores obtidos, trazendo uma linha reta com poucos picos.

### Histograma

```python
# Show the histogram for Tc
Experiment.show_histogram(experiment_1_obj, "Experiment 1")

# Show the histogram for Tv
Experiment.show_histogram(experiment_2_obj, "Experiment 2", "orange")

# Show both histograms side by side
Experiment.show_histograms_side_by_side(experiment_1_obj, experiment_2_obj, "Experiment 1 vs Experiment 2", ["Experiment 1", "Experiment 2"])
```

> Com o histogram sobreposto é possível de se observar que o experimento 2 grande parte de seus valores a esquerda do histograma do experimento 1, assim como já foi percebida a diferença em relação ao gráfico em barras.

### Teste de Variabilidade

```python
Experiment.show_variability_plot(experiment_1_obj, "Experiment 1")
Experiment.show_variability_plot(experiment_2_obj, "Experiment 2", "orange")
```

```python
Experiment.show_variability(experiment_1_obj, "Variability - Experiment 1")
Experiment.show_variability(experiment_2_obj, "Variability - Experiment 2")
```

> Com o teste de variabilidade é possível se observar que o resultado obtido pelo experimento possui uma boa qualidade, visto que o ambiente de execução possuiu uma variabilidade abaixo de .15, mais precisamente abaixo de 0.027 em ambos os experimentos.

### Média e Desvio padrão

```python
Experiment.show_mean_deviation_plot(experiment_1_obj, "Experiment 1", 0.0025)
Experiment.show_mean_deviation_plot(experiment_2_obj, "Experiment 2", 0.0025)
```

> O gráfico da média e desvio padrão demonstra um grande afastamento entre as duas "linhas" geradas tanto do experimento 1 quanto do experimento 2 e os valores segue o padrão visualizado nos gráficos anteriores, com o experimento 1 com valores em torno de .045 e o experimento 2 com valores próximos a .4. Os valores parecem bem estabilizados, porém só com o gráfico anterior, de variabilidade, é possível ver o comportamento que só se estabiliza próximos dos 1000 experimentos.

### Gráfico de disperção

```python
Experiment.show_scatter_plot(experiment_1_obj, 'b', "Experiment 1")
Experiment.show_scatter_plot(experiment_2_obj, 'r', "Experiment 2")

Experiment.show_scatter_plot_side_by_side(experiment_1_obj, experiment_2_obj, 'b', 'r', "const. exp 1", "const. exp 2","Experiment 1 vs Experiment 2")
```

> Na sobreposição de ambos os gráficos de disperção, fica clara a diferença de tempo e a distribuição desses dados.

### Comparação da média

```python
import numpy as np
from IPython.display import Markdown, display

mean1 = np.mean(experiment_1_obj.mean)
mean2 = np.mean(experiment_2_obj.mean)

if(mean1 > mean2):
    display(Markdown("<h4>Experimento 1 possui a maior média de tempo</h4>"))
    mean_difference_percent = (mean1 - mean2)/mean1 * 100
else:
    display(Markdown("<center><h4>Experimento 2 possui a maior média de tempo</h4></center>"))
    mean_difference_percent = (mean2 - mean1)/mean2 * 100

print("{:.4f}%".format(mean_difference_percent))
```

## Conclusão
> Com base nesse experimento, podemos observar que existe uma variação entre o tempo de execução do experimento 1 e 2. Agora, a diferença já ficou bem maior do que foi visto anteriormente na comparação que variava apenas o simbolo Maior que e Menor que. Assim, ficando em torno de 8% de tempo de processamento maior do experimento 1 em relação ao 2.
