```python
import numpy as np
import matplotlib.pyplot as plt


class KMeans: # (0, 0, 0, 0)
    def __init__(self, k, max_iters=100): # (1, 0, 0, 2)
        self.k = k # (0, 0, 0, 1)
        self.max_iters = max_iters # (0, 0, 0, 1)

    def initialize_centroids(self, X):  # (1, 0, 0, 1)
        centroids_indices = np.random.choice(X.shape[0], self.k, replace=False)  # (0, 1, 0, 3)
        centroids = X[centroids_indices]  # (0, 0, 0, 2)
        return centroids  # (0, 0, 0, 1)
    
    def assign_clusters(self, X, centroids):  # (1, 0, 0, 1)
        distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))  # (0, 3, 0, 10)
        return np.argmin(distances, axis=0)  # (0, 0, 0, 2)
    
    def update_centroids(self, X, clusters):  # (1, 0, 0, 1)
        centroids = np.zeros((self.k, X.shape[1]))  # (0, 1, 0, 4)
        for i in range(self.k):  # (0, 0, 0, 2)
            centroids[i] = np.mean(X[clusters == i], axis=0)  # (0, 2, 0, 7)
        return centroids  # (0, 0, 0, 1)
    
    def fit(self, X):  # (1, 0, 0, 1)
        centroids = self.initialize_centroids(X)  # (0, 0, 0, 3)
        for _ in range(self.max_iters):  # (0, 0, 0, 2)
            clusters = self.assign_clusters(X, centroids)  # (0, 0, 0, 3)
            new_centroids = self.update_centroids(X, clusters)  # (0, 0, 0, 3)
            if np.all(centroids == new_centroids):  # (0, 3, 0, 8)
                break  # (0, 0, 0, 1)
            centroids = new_centroids  # (0, 0, 0, 2)
        return clusters, centroids  # (0, 0, 0, 2)


# Gerando dados de exemplo
np.random.seed(0)
X = np.random.rand(100, 2)

# Criando e executando o algoritmo KMeans
kmeans = KMeans(k=3)
clusters, centroids = kmeans.fit(X)

# Plotando os clusters e os centroides
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='red')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
