```python
import numpy as np
from time import time
from matplotlib import pyplot as plt

r = 1000        # Número de repetições do experimento
R = range(r)    # Vetor das repetições

T = []          # Resultado de cada experimento
n = 10 ** 6     # Quantidade do dado
N = range(n)    # Vetor dos dados 'n'

V = []          #Vetor da variabilidade
M = []          #Vetor da media
S = []          #Vetor do Desvio padrao

for j in R:

    tic = time()

    for i in N:
        a = 1

    tac = time()
    
    T.append(tac-tic)
    
    S.append(np.std(T))
    M.append(np.mean(T))
    
print("Terminou!")
```

```python
plt.plot(M)
plt.plot(S)
plt.grid()
plt.legend(["Mean", "Desvio"])
plt.show()
```
