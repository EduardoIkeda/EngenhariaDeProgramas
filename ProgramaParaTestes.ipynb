# Programa base para realização de experimentos

## Código Base

```python
import numpy as np
from time import time
from matplotlib import pyplot as plt

r = 1000        # Número de repetições do experimento
R = range(r)    # Vetor das repetições

T = []          # Resultado de cada experimento
n = 10 ** 6     # Quantidade do dado
N = range(n)    # Vetor dos dados 'n'

V = []          #Vetor da variabilidade
M = []          #Vetor da media
S = []          #Vetor do Desvio padrao

for j in R:

    tic = time()

    for i in N:
        a = 1

    tac = time()
    
    T.append(tac-tic)
    
    S.append(np.std(T))
    M.append(np.mean(T))
    V.append(np.std(T)/np.mean(T))
    
print("Terminou!")
```

## Gráficos

```python
plt.bar(R,T)
plt.show()
```

```python
plt.plot(T, color='red')
plt.bar(R,T)
plt.show()
```

```python
plt.hist(T, 20)
plt.xlabel('tempo')
plt.ylabel('frequencia')
plt.grid()
plt.show()
#Analise qualitativa dos dados, vendo se há uma distribuição normal dos dados
```

```python
S = np.std(T)
M = np.mean(T)
Variabilidade = S/M

print(Variabilidade)
#Calculo da variabilidade para checar se a maquina é estavel para os experimentos. Idealmente precisa estar abaixo de .15
```
