```python
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt

class KMeans:
    def __init__(self, K, max_iter=100):
        self.K = K
        self.max_iter = max_iter

    def fit(self, X):
        n_samples, n_features = X.shape
        self.centroids = X[np.random.choice(n_samples, self.K, replace=False)]

        for _ in range(self.max_iter):
            clusters = [[] for _ in range(self.K)]
            for sample_index, sample in enumerate(X):
                closest_centroid_index = np.argmin(np.linalg.norm(sample - self.centroids, axis=1))
                clusters[closest_centroid_index].append(sample_index)

            new_centroids = np.zeros((self.K, n_features))
            for cluster_index, cluster in enumerate(clusters):
                if cluster:
                    new_centroids[cluster_index] = np.mean(X[cluster], axis=0)

            if np.allclose(self.centroids, new_centroids):
                break

            self.centroids = new_centroids

        self.labels_ = np.zeros(n_samples, dtype=int)
        for cluster_index, cluster in enumerate(clusters):
            self.labels_[cluster] = cluster_index

    def predict(self, X):
        closest_centroid_indices = np.argmin(np.linalg.norm(X - self.centroids[:, np.newaxis], axis=2), axis=0)
        return closest_centroid_indices

# Diretrio onde esto as imagens
INPUT_PATH = 'C:\\Users\\duike\\OneDrive\\Documentos\\Repositorios\\EngenhariaDeProgramas\\FACES'

# Funo para carregar as imagens do diretrio especificado
def load_images(input_path, num_images, as_gray=True):
    images = []
    filenames = os.listdir(input_path)
    for filename in filenames[:num_images]:
        filepath = os.path.join(input_path, filename)
        try:
            image = Image.open(filepath)
            if as_gray:
                image = image.convert('L')
            images.append(image)
        except:
            print(f"Erro ao carregar a imagem: {filename}")
    return images

# Funo para plotar o grfico de frequncia das intensidades de cores de cada cluster
def plot_cluster_histograms(images, labels):
    num_clusters = len(np.unique(labels))
    for i, image in enumerate(images):
        fig, axes = plt.subplots(1, num_clusters + 1, figsize=(15, 5))
        for j, label in enumerate(np.unique(labels)):
            cluster_indices = np.where(labels == label)[0]
            cluster_pixels = np.array([image.getpixel((idx % image.width, idx // image.width)) for idx in cluster_indices])
            axes[j].hist(cluster_pixels, bins=64, color='gray', alpha=0.5)
            axes[j].set_title(f'Cluster {label}')
        axes[num_clusters].imshow(image, cmap='gray')
        axes[num_clusters].set_title('Original Image')
        plt.show()

# Quantidade de imagens que voc deseja carregar
num_images = 1

# Carrega as imagens do diretrio
images = load_images(INPUT_PATH, num_images)

# Converte as imagens para arrays numpy e achat-las
flattened_images = [np.array(image).reshape(-1, 1) for image in images]

# Suponha que 'flattened_images' seja uma lista de arrays numpy representando as imagens achatadas

kmeans = KMeans(K=4)  # Define o nmero de clusters desejado
for i, flattened_image in enumerate(flattened_images):
    kmeans.fit(flattened_image)  # Aplica o algoritmo KMeans
    labels = kmeans.labels_  # Obtm os rtulos dos clusters
    clustered_image = kmeans.centroids[labels].reshape(images[i].size)  # Reconstri a imagem
    clustered_image = clustered_image.astype(np.uint8)
    clustered_image = Image.fromarray(clustered_image)
    clustered_image.show()

    # Plotar grfico de frequncia das intensidades de cores de cada cluster
    plot_cluster_histograms([images[i]], labels)
```
